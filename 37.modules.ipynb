```python
# module = a file containing python code, May Contain functions,classes ,etc
# used with (modular programming)-> which is to seperate a program into parts


import message_module_37 # this is a module

message_module_37.hello() # hello is a function inside message_module_37 module
message_module_37.bye()

```

```python
# we can give our module an alias name(nick name)

import message_module_37 as m37  
m37.bye()
m37.hello()
```

```python
# ANOTHER WAY OF IMPORTING A MODULE : 

# from (module) import (functions,classes) you want to import
from message_module_37 import hello,bye

# now we no longer need module name ,while calling function
hello()
bye()
```

```python
# ANOTHER WAY OF IMPORTING A MODULE : 

# from (module) import import(everything)
from message_module_37 import *

# this way of importing is not recommended if you are working with large file or large project 
    #- containing many modules ,because you can run into a naming conflict , becuase some of these modules
        #- may have same function,class,variable name 

# Note : It is much more safer to write name of the module and then the name of the function 
    #-eg   m37.hello()       

```
